```python
def add_spaces(text):
    text = text.strip()

   
    
    result = re.sub(r'([A-Z]+)(\d)', r'\1 \2', text)
    result = re.sub(r'(\d+)([A-Z])', r'\1 \2', result)
    
    replacements = {
        'DWELLINGTYPE': 'DWELLING TYPE',
        'BEDROOMTYPE': 'BEDROOM TYPE',
        'RENTALUNITAMOUNT': 'RENTAL UNIT AMOUNT',
        'POPULATIONAMOUNT': 'POPULATION AMOUNT',
        'AVERAGERENT': 'AVERAGE RENT',
        'CONSTRUCTEDBEFORE': 'CONSTRUCTED BEFORE',
        'CONSTRUCTEDAFTER': 'CONSTRUCTED AFTER',
        'REFYEAR': 'YEAR'
    }
    
    for old, new in replacements.items():
        result = result.replace(old, new)
    
    return result

data.columns = [add_spaces(col) for col in data.columns]
print(data.columns.tolist())
```
